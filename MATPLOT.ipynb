```python
import matplotlib.pyplot as plt
print("Successfully installed")
x=[1,2,3,4]
y=[10,20,30,40]

plt.plot(x,y)
plt.title("Graph of X against Y")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
#Example 2

students=['Honour', 'Treasure', 'Charity', 'Faidat']
Scores=[85,90,78,92]


Colours=['green', 'yellow', 'blue', 'brown']
plt.bar(students, Scores,  color=Colours)
plt.title("Students scores")
plt.xlabel("Student")
plt.ylabel("Scores")
plt.show()
```

```python
height=[150,160,165,170, 175]
weight=[50,55,60,65,70]

plt.scatter(height, weight, color="Purple")
plt.title("Graph of Height Against Weight")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

```python
Labels=["Python", "Java", "Analysis", "Science"]
Sizes=[40,25,20,15]

color = ['green', 'yellow', 'blue', 'brown']
plt.pie(Sizes, labels=Labels,autopct="%1.0f%%",colors=color) #autopct is for percentage values to show
plt.title("GRAPH OF LANGUAGES")
plt.show()
```
