```python
#Example 2

data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David"],
    "Age": [25, None, 30, 25, 20],
    "City": [" Lagos ", "Abuja", "lagos", " Lagos ", None],
    "Gender": ["F", "M", "Male", "F", "M"]
}
df = pd.DataFrame(data)
print("Data Set:\n", df)

print("Data Set:\n", df.info())

#finding missing details
print("Missing Data Set:\n", df.isnull().sum())

#fill age with mean
df['Age']=df['Age'].fillna(df['Age'].mean())
print("Filling the missing age in the Data Set:\n", df)

#Change age with mean
df['Age']=df['Age'].astype(int)
print("Change to Integer:\n", df)

#fill mising City 
df['City']=df['City'].fillna('Unknown')
print("Filling the missing City in the Data Set:\n", df)

#To remove spaces (bar space during tying in between the words)
df['City']=df['City'].str.strip() 
print('New Data Set with no space:\n', df)

#To put all the Cities on Lower Case
df['City']=df['City'].str.lower() 
print('New Data Set with no space:\n', df)

#To put all the Cities on Upper Case
df['City']=df['City'].str.upper() 
print('New Data Set with no space:\n', df)

#To put all the Cities on Title Case
df['City']=df['City'].str.title() 
print('New Data Set with no space:\n', df)

#Removing Duplicate
df=df.drop_duplicates()
print("Removed Duplicate Data Set", df)

#Standardizing Gender values 
df['Gender']=df['Gender'].replace({
    "Male": "M", 
    "Female": "F"
})   #Dictionary 
print("Replaced the Gender to be uniform:\n", df)


```

```python
import pandas as pd
import numpy as np


Joy = {
    'OrderID': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    'Date': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-03', '2025-01-04', 
             '2025-01-05', '2025-01-06', '2025-01-07', '2025-01-08', '2025-01-09'],
    'CustomerName': ['John Doe', 'Jane Smith', None, 'Alice Green', 'Bob White', 
                     None, 'Mike Brown', 'Sarah Black', 'Emma White', 'James King'],
    'ProductCategory': ['Electronics', 'Home Goods', 'Clothing', 'Electronics', None, 
                        'Furniture', 'Home Goods', 'Clothing', None, 'Electronics'],
    'Quantity': [2, 1, 3, 1, 5, 1, 2, 4, 3, None],
    'Price': [300, 50, 25, 150, 100, 500, None, 30, 80, 200],
    'Total': [600, 50, 75, 150, 500, 500, None, 120, 240, 200],
    'PaymentMethod': ['Credit Card', 'PayPal', 'Debit Card', None, 'Credit Card', 
                      'PayPal', 'Debit Card', 'Credit Card', None, 'Credit Card'],
    'DeliveryStatus': ['Delivered', 'Pending', 'Delivered', 'Canceled', 'Delivered', 
                       None, 'Shipped', 'Pending', 'Delivered', 'Delivered']
}


print("====================DISPLAY OF THE DATA SET=======================")
df=pd.DataFrame(Joy)
print("Data Set:\n", df)

print("\n====================ATTRIBUTES OF THE DATA SET=======================")
print("Data Shape\n", df.shape)
print("Data Column:\n", df.columns)

print("\n====================METHODS OF THE DATA SET=======================")
print("Data Head:\n", df.head())
print("\n Data Tail:\n", df.tail())
print("\n Data Info:\n", df.info())
print("\n Data Sample:\n", df.sample())
print("\n Data Description:\n", df.describe())
print("\n Missing Data:\n", df.isnull())
print("\n Total Number of Missing Data:\n", df.isnull().sum())

print("\n==================== CLEANING AND ANALYSING THE DATA SET=======================")

print("\n==================== Filling the Mising Names with No Name=======================")
df["CustomerName"]=df["CustomerName"].fillna("No name")
print("Filling Missing Name:\n", df)

print("\n==================== Filling the Mising or inconsistent Products with Miscelleneous=======================")
df["ProductCategory"]=df["ProductCategory"].fillna("Misceleneous")
print("Filling Missing Name:\n", df)

print("\n==================== Filling the Mising Values or quantities with 'mean' values =======================")
df["ProductCategory"]=df["ProductCategory"].fillna("Misceleneous")
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Price"] = df["Price"].fillna(df["Price"].mean())
print("Filled Values and Quanities:\n", df)

print("\n==================== Filling the Mising Values or quantities with 'mean' values and calulate the total=======================")
df["Total_Price"] = df["Price"].sum()
df["Total Quanties"]=df["Quantity"].sum()
print("Total Values and Quanities:\n", df)

print("\n==================== Filling the Payment Method missing data with 'Credit Card'=======================")
df["PaymentMethod"]=df["PaymentMethod"].fillna("Credit Card")
print("Missing Payment Method-show 'Credit Card':\n",df)

print("\n==================== Filling the missing or incorrect Delivery Statuses and fill with 'Pending'=======================")
df["DeliveryStatus"]=df["DeliveryStatus"].fillna("Pending")
print("Missing Dilivery Statuses-show 'Credit Card':\n",df)

print("\n==================== Show the Cleaned Data Set=======================")

print("\n==================== Show the row where order_Id is 1004=======================")
Customer_1004=df[df["OrderID"] == 1004]
print("Customer ID_1004:\n",Customer_1004)

print("\n==================== Total Sales Per Product=======================")
df["Sales"] = df["Price"] * df["Quantity"]
total_Sales_of_each_Product = df.groupby("ProductCategory")["Sales"].sum()
print("Each Product Total Sales\n",total_Sales_of_each_Product)


print("\n==================== Products Purchased with Credit Card=======================")
products_purchased_Credit_Card=df[df["PaymentMethod"]=="Credit Card"]
print("Products Purchased by Credit Card:\n",products_purchased_Credit_Card)

print("\n==================== Products Purchased with Credit Card and Delivered=======================")
products_purchased_Credit_Card_Delivered = df[(df["PaymentMethod"]=="Credit Card") & (df["DeliveryStatus"]=="Delivered")]
print("Products Purchased by Credit Card and Delivered:\n",products_purchased_Credit_Card_Delivered)




```
