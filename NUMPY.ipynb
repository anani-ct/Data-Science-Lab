```python
import numpy as np

Students_Scores = np.array([[65,70,80],
                            [45,55,60],
                            [90,95,85],
                            [70,60,75],
                            [50,40,65]])

# 1. Select all rows where Test1 > 60
rows_test1_gt_60 = Students_Scores[Students_Scores[:,0] > 60]
print("Test1 > 60:\n", rows_test1_gt_60)

# 2. Select all rows where Test1 <= 60
rows_test1_le_60 = Students_Scores[Students_Scores[:,0] <= 60]
print("Test1 <= 60:\n", rows_test1_le_60)

# 3. Select rows where Test1 > 60 AND Test3 >= 75
rows_condition1 = Students_Scores[(Students_Scores[:,0] > 60) & 
                                   (Students_Scores[:,2] >= 75)]
print("Test1 > 60 and Test3 >= 75:\n", rows_condition1)

# 4. Select rows where Test1 < 70 AND Test2 < 60
rows_condition2 = Students_Scores[(Students_Scores[:,0] < 70) & 
                                   (Students_Scores[:,1] < 60)]
print("Test1 < 70 and Test2 < 60:\n", rows_condition2)

# 5. Replace all scores in Test3 that are less than 70 with 70
Students_Scores[Students_Scores[:,2] < 70, 2] = 70
print("Updated Students_Scores:\n", Students_Scores)
```

```python
import numpy as np
Students = np.array([
"Moyo", "Charity", "Honour", "Ifeoluwa",
"Treasure", "Itunuoluwa", "David", "Grace"
])

Subjects = np.array([
"Mathematics", "English", "Agric", "Computer"
])

Scores = np.array([
[78, 65, 80, 90],
[85, 70, 75, 88],
[60, 75, 70, 72],
[90, 88, 92, 95],
[55, 60, 58, 65],
[72, 68, 74, 80],
[88, 82, 85, 90],
[70, 72, 78, 76]
])

print("Students:\n", Students)
print("Subjects:\n", Subjects)
print("Scores:\n", Scores)

Shape_Array=np.shape(Scores)
print("Shape:", Shape_Array)

import numpy as np

# Student names
Students = np.array([
"Moyo", "Charity", "Honour", "Ifeoluwa",
"Treasure", "Itunuoluwa", "David", "Grace"
])

# Subjects
Subjects = np.array([
"Mathematics", "English", "Agric", "Computer"
])

# Scores dataset
Scores = np.array([
[78, 65, 80, 90],
[85, 70, 75, 88],
[60, 75, 70, 72],
[90, 88, 92, 95],
[55, 60, 58, 65],
[72, 68, 74, 80],
[88, 82, 85, 90],
[70, 72, 78, 76]
])

# 1. Total score for each student
total_scores = Scores.sum(axis=1)
print("Total Scores:", total_scores)

# 2. Average score for each student
average_scores = Scores.mean(axis=1)
print("Average Scores:", average_scores)

# 3. Top performing student
top_student_index = np.argmax(total_scores)
top_student = Students[top_student_index]
print("Top Student:", top_student)

# 4. Lowest performing student
lowest_student_index = np.argmin(total_scores)
lowest_student = Students[lowest_student_index]
print("Lowest Student:", lowest_student)

# 5. Most difficult subject (lowest average)
subject_avg = Scores.mean(axis=0)
difficult_subject = Subjects[np.argmin(subject_avg)]
print("Most Difficult Subject:", difficult_subject)

# 6. Best performing subject (highest average)
best_subject = Subjects[np.argmax(subject_avg)]
print("Best Performing Subject:", best_subject)

# 7. Highest score in dataset
highest_score = Scores.max()
print("Highest Score:", highest_score)

# 8. Lowest score in the dataset
lowest_score = Scores.min()
print("Lowest Score:", lowest_score)

# 9. Overall class average
class_average = Scores.mean()
print("Class Average:", class_average)

# 10. Students who passed all subjects (>=70)
passed_all = Students[np.all(Scores >= 70, axis=1)]

print("Students who passed all subjects:", passed_all)

# 11. Students who failed at least one subject
failed_one = Students[np.any(Scores < 70, axis=1)]
print("Students who failed at least one subject:", failed_one)

# 12. Students scoring above 80 in Computer
computer_above_80 = Students[Scores[:,3] > 80]
print("Students scoring above 80 in Computer:", computer_above_80)

# 13. Student with highest Mathematics score
math_top = Students[np.argmax(Scores[:,0])]
print("Top Mathematics Student:", math_top)

# 14. Students above class average
above_avg_students = Students[average_scores > class_average]
print("Students above class average:", above_avg_students)
```
